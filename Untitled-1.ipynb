```python
import os
import shutil

# define the source directory where the files are located
source_dir = "path/to/source/directory"

# define the destination directories for each class
class_dirs = {
    'bag',
    'big_cardboard',
    'bulky',
    'compactor',
    'cylinders',
    'dumpster',
    'electrical_and_electronic _waste',
    'garbage_bag',
    'glass',
    'green_waste',
    'hygiene',
    'plastic_bag',
    'recycle_waste',
    'textile'
}

# create the destination directories if they don't exist
for class_dir in class_dirs.values():
    if not os.path.exists(class_dir):
        os.makedirs(class_dir)

# loop through the files in the source directory
for filename in os.listdir(source_dir):
    # get the class of the file (assuming it's in the filename)
    file_class = filename.split("_")[0]  # adjust this to match your naming convention
    
    # get the destination directory for the class
    class_dir = class_dirs.get(file_class)
    
    # move the file to the destination directory
    shutil.copy2(os.path.join(source_dir, filename), os.path.join(class_dir, filename))
```
